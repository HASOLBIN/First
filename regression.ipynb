# 회귀분석

- 비선형을 선형으로
- 돈과 관련된 것은 log를 취하여 직선화시켜야함(선형관계로 바꾸기)
- 비선형인 데이터들은 변환을 통하여 선형관계로 바꾼 후 회귀분석을 실시하여야 한다.

```python
import numpy as np 
import pandas as pd #기초 함수 불러오기

x = np.random.randn(3) #랜덤한 숫자 3개 뽑아내기
x
```

```python
y=x*100+10
# w=100(기울기), b=10(절편)이라고 할 때 이 값을 컴퓨터가 찾아낼 수 있는가?
y
```

```python
lr=1.0 # 시스템마다의 학습 속도(예를들어 엑셀을 밟는 정도, 시스템마다 반응 속도가 다르기때문에 최적의 lr을 찾아야 함)
w=np.random.rand() 
b=np.random.rand()
w,b
# w와 b에 초기값을 지정 한 후 이제 학습을 통해 컴퓨터가 w와 b를 찾아낼 수 있는지 보자
# 내가 예측한 값이 실제 값과 얼마나 차이가 있는지 보고 그 차이(error)를 줄여가면서 w와 b를 찾는다.
```

```python
n_epoch=300 
for epoch in range(n_epoch): #300번 돌려볼것이다
    y_pred = w*x + b
    error = np.abs(y - y_pred).mean() #이 error가 작으면 작을수록 좋은 것
    
    if error < 0.01:
        break # error가 0.01보다 작으면 for문을 빠져나온다 = 최적화시킨다
        
    w = w - lr*((y_pred-y)*x).mean() # abs는 절댓값 x를 곱해주기때문에 오류의 변환은 x에 영향을 미친다.
    # 뮤는 lr Grad는 기울기 : (y_pred-y)*x mean은 x를 3개로 받았기 때문에 그 평균값을 곱해줌
    b = b - lr*(y_pred-y).mean() # b는 절편이기 때문에 x를 곱해주지 않음
    if epoch%10 ==0: 
        print('epoch=', epoch, 'error=', error, 'w=',w,'b=',b) # epoch, error, w, b를 출력한다

print('--------------------------------------------------------------------------------')    
print('epoch=', epoch, 'error=', error, 'w=',w,'b=',b) # epoch, error, w, b를 출력한다

# 처음에 x(3개의 랜덤한 수)만 주어졌는데 컴퓨터가 학습을 통해 오차(error)를 줄여나가면서 w와 b가 100과 10임을 찾아냈음
```
